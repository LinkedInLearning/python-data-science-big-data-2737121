# Operaciones y transformaciones en Pandas

```python
df['TOTAL_DELAY'] = df['DEPARTURE_DELAY'] + df['ARRIVAL_DELAY']
```

```python
df
```

```python
df['TIEMPO_GANADO'] = df['SCHEDULED_TIME'] - df['ELAPSED_TIME']
```

```python
df['RETRASO_BINARIO'] = df['ARRIVAL_DELAY'] > 15
```

```python
df['DIA_SEMANA'] = df['DAY_OF_WEEK'].map({
    1: 'Lunes', 2: 'Martes', 3: 'MiÃ©rcoles',
    4: 'Jueves', 5: 'Viernes', 6: 'SÃ¡bado', 7: 'Domingo'
})
```

```python
df['RATIO_RETRASO'] = df['ARRIVAL_DELAY'] / df['DISTANCE']
```

```python
df['RATIO_RETRASO'] = df['RATIO_RETRASO'].round(2)
```

```python
df['TIPO_VUELO'] = df['DISTANCE'].apply(lambda x: 'Corto' if x < 500 else 'Largo')
```

```python
def clasificar_hora(salida):
    if salida < 600:
        return 'Madrugada'
    elif salida < 1200:
        return 'MaÃ±ana'
    elif salida < 1800:
        return 'Tarde'
    else:
        return 'Noche'
```

```python
df['FRANJA_HORARIA'] = df['SCHEDULED_DEPARTURE'].apply(clasificar_hora)
```

```python
columnas_a_eliminar = ['RATIO_RETRASO']
try:
    df = df.drop(columns=columnas_a_eliminar)
except:
    pass
```
